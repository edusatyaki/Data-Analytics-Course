# 3. Data Cleaning - customer_id

# 4. Data Cleaning - customer_name

**Step 1**: Count the NAN values

```python
df['customer_name'].isnull().sum()
```

**Step 2**: Replace NAN with 'Unknown' and ensure Proper Case

```python
df['customer_name'] = df['customer_name'].fillna("Unknown")
df['customer_name'] = df['customer_name'].str.strip().str.title()
df['customer_name'].unique()[:10]
```

# 5. Data Cleaning - age

**Step 1**: Check the value range (we found invalid values like -7 and 999)

```python
print('Min:', df['age'].min(), 'Max:', df['age'].max())
df['age'].isnull().sum()
```

**Step 2**: Replace invalid ages (age < 0 or age > 100) with NAN

```python
import numpy as np
df.loc[(df['age'] < 0) | (df['age'] > 100), 'age'] = np.nan
```

**Step 3**: Fill NAN with the median age

```python
median_age = df['age'].median()
df['age'] = df['age'].fillna(median_age)
print('Median used:', median_age)
df['age'].isnull().sum()
```

**Step 4**: Convert data type from Float to Int

```python
df['age'] = df['age'].astype(int)
df['age'].dtypes
```

# 6. Data Cleaning - gender

Check the unique values - they have case/short-form issues (M, m, MALE, f, FEMALE, etc.)

```python
df['gender'].unique()
```

**Step 1**: Standardise text to a single clean form (Male / Female / Other)

```python
df['gender'] = df['gender'].str.strip().str.upper()

gender_map = {
    'M': 'Male', 'MALE': 'Male',
    'F': 'Female', 'FEMALE': 'Female',
    'OTHER': 'Other'
}
df['gender'] = df['gender'].map(gender_map)
df['gender'].unique()
```

**Step 2**: Fill NAN with 'Other'

```python
df['gender'] = df['gender'].fillna("Other")
df['gender'].isnull().sum()
```

**Step 3**: Convert into Int by mapping  Male -> 1, Female -> 2, Other -> 3

```python
gender_code = {'Male': 1, 'Female': 2, 'Other': 3}
df['gender'] = df['gender'].map(gender_code).astype(int)
df['gender'].unique()
```

# 7. Data Cleaning - region

**Step 1**: Count the NAN values

```python
df['region'].isnull().sum()
```

**Step 2**: Fill NAN with 'Unknown' and standardise case

```python
df['region'] = df['region'].fillna("Unknown").str.strip().str.title()
df['region'].unique()
```

**Step 3**: Convert region into short codes (Central -> C, West -> W, North -> N, East -> E, South -> S)

```python
region_map = {
    'Central': 'C', 'West': 'W', 'North': 'N',
    'East': 'E', 'South': 'S', 'Unknown': 'U'
}
df['region'] = df['region'].map(region_map)
df['region'].unique()
```
