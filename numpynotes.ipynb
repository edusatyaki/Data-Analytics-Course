## 1. Array Creation
* `np.array()`: Create from list.

* `np.ones()`: Array of ones.

* `np.arange()`: Range of values.
* `np.linspace()`: Linearly spaced.
* `np.eye()`: Identity matrix.
* `np.identity()`: Square identity.
* `np.empty()`: Uninitialized.
* `np.tri()`: Triangle 1s.

```
# Import NumPy library (required to use np functions)
import numpy as np

# Create a NumPy array from a Python list [1, 2, 3] and print it
print('array:\n', np.array([1, 2, 3]))

# Create a 2x2 NumPy array filled with zeros and print it
print('\nzeros:\n', np.zeros((2, 2)))

# Create a 2x2 NumPy array filled with ones and print it
print('\nones:\n', np.ones((2, 2)))

# Create a 2x2 NumPy array filled with the value 7 and print it
print('\nfull:\n', np.full((2, 2), 7))

# Create an array starting from 0 to 10 (exclusive) with step size 2
print('\narange:\n', np.arange(0, 10, 2))

# Create 5 evenly spaced values between 0 and 1 and print them
print('\nlinspace:\n', np.linspace(0, 1, 5))

# Create a 2x2 identity matrix (diagonal elements = 1, others = 0) and print it
print('\neye:\n', np.eye(2))

# Create a 2x2 identity matrix using identity() function and print it
print('\nidentity:\n', np.identity(2))

# Create a 2x2 array with uninitialized (random garbage)
# values from memory and print it
print('\nempty:\n', np.empty((5, 5)))

# Create a 3x3 lower triangular matrix filled with ones and print it
print('\ntri:\n', np.tri(3))
```

## 3. Reshaping
* `np.reshape()`: Change shape.
* `.flatten()`: Collapse to 1D.
* `.ravel()`: Flatten view.
* `.T`: Transpose.
* `np.resize()`: Resize array.
* `np.expand_dims()`: Add dimension.
* `np.squeeze()`: Remove 1-dims.
* `np.transpose()`: Permute dimensions.
* `np.swapaxes()`: Interchange two axes.
* `np.rollaxis()`: Roll axis.

```
# Import NumPy library
import numpy as np

# Create a 1D array with values from 0 to 5
x = np.arange(6)
print(x)
# Create a 2x2 array
y = np.array([[1, 2], [3, 4]])
print(y)
# Reshape the 1D array x into a 2x3 array
print('reshape:\n', np.reshape(x, (2, 3)))

# Flatten the 2D array into a 1D copy of the array
print('\nflatten:', y.flatten())

# Convert the 2D array into a flattened 1D view (no copy if possible)
print('ravel:', y.ravel())

print(y)
# Transpose the array (swap rows and columns)
print('\nT (Transpose):\n', y.T)

# Resize array x into a 2x2 array (repeats elements if needed)
print('\nresize:\n', np.resize(x, (2, 2)))

# Add a new dimension at axis 0 and print the new shape
print('\nexpand_dims shape:', np.expand_dims(x, axis=0).shape)

# Create a 3D array with one row and two elements
z = np.array([[[1, 2]]])

# Remove single-dimensional entries from the shape
print('squeeze shape:', np.squeeze(z).shape)

# Transpose array using transpose() function
print('\ntranspose:\n', np.transpose(y))

# Swap axis 0 and axis 1 of array y
print('\nswapaxes:\n', np.swapaxes(y, 0, 1))

# Move axis 2 to position 0 and print the resulting shape
print('\nrollaxis shape:', np.rollaxis(np.ones((3,4,5)), 2, 0).shape)
```

## 5. Statistics
* `np.mean()`: Mean.
* `np.median()`: Median.
* `np.std()`: Std Deviation.
* `np.var()`: Variance.
* `np.sum()`: Sum.
* `np.min()`: Minimum.
* `np.max()`: Maximum.
* `np.argmin()`: Index min.
* `np.argmax()`: Index max.
* `np.percentile()`: Percentile.
* `np.corrcoef()`: Correlation.
* `np.cov()`: Covariance.
* `np.cumsum()`: Cumulative sum.
* `np.cumprod()`: Cumulative prod.

```
# Import NumPy library
import numpy as np

# Create a 2x2 NumPy array for statistical operations
stat_arr = np.array([[1, 2], [3, 4]])

# Calculate the mean (average) of all elements
print('mean:', np.mean(stat_arr))

# Calculate the median (middle value) of all elements
print('median:', np.median(stat_arr))

# Calculate the standard deviation of the elements
print('std:', np.std(stat_arr))

# Calculate the variance of the elements
print('var:', np.var(stat_arr))

# Calculate the sum of all elements
print('sum:', np.sum(stat_arr))

# Find the minimum value in the array
print('min:', np.min(stat_arr))

# Find the maximum value in the array
print('max:', np.max(stat_arr))

# Find the index of the minimum value (flattened array index)
print('argmin:', np.argmin(stat_arr))

# Find the index of the maximum value (flattened array index)
print('argmax:', np.argmax(stat_arr))

# Calculate the 50th percentile (same as median)
print('percentile (50th):', np.percentile(stat_arr, 50))

# Calculate correlation coefficient matrix between rows
print('\ncorrcoef:\n', np.corrcoef(stat_arr))

# Calculate covariance matrix between rows
print('\ncov:\n', np.cov(stat_arr))

# Calculate cumulative sum of elements (flattened order)
print('\ncumsum:', np.cumsum(stat_arr))

# Calculate cumulative product of elements (flattened order)
print('cumprod:', np.cumprod(stat_arr))
```

## 6. Searching
* `np.where()`: Conditional select.
* `np.searchsorted()`: Find insertion index.
* `np.nonzero()`: Non-zero indices.
* `np.argwhere()`: Indices of condition.
* `np.extract()`: Extract by condition.

```
# Import NumPy library
import numpy as np

# Create a NumPy array for searching operations
search_arr = np.array([10, 20, 30, 40])

# Find indices where elements are greater than 20
print('where > 20:', np.where(search_arr > 20))

# Find the index where value 25 should be inserted to maintain sorted order
print('searchsorted (insert 25):', np.searchsorted(search_arr, 25))

# Find indices of non-zero (True) elements for condition > 20
print('nonzero:', np.nonzero(search_arr > 20))

# Return indices of elements greater than 20 in row-column format
print('\nargwhere > 20:\n', np.argwhere(search_arr > 20))

# Extract elements greater than 20 from the array
print('\nextract > 20:', np.extract(search_arr > 20, search_arr))
```

## 7. Sorting
* `np.sort()`: Return sorted.
* `np.argsort()`: Sort indices.
* `np.lexsort()`: Indirect sort.
* `np.partition()`: Partial sort.
* `np.argpartition()`: Partition indices.

```
# Import NumPy library
import numpy as np

# Create a NumPy array for sorting operations
sort_arr = np.array([3, 1, 2, 5, 4])

# Sort the array in ascending order
print('sort:', np.sort(sort_arr))

# Return the indices that would sort the array
print('argsort:', np.argsort(sort_arr))

# Perform indirect sorting using multiple keys (primary and secondary sorting)
print('lexsort:', np.lexsort((sort_arr, [0, 1, 0, 1, 0])))

# Rearrange the array so that the element at index 2 is in sorted position
# Elements before it are smaller and after it are larger (not fully sorted)
print('partition (k=2):', np.partition(sort_arr, 2))

# Return indices that would partition the array at index 2
print('argpartition (k=2):', np.argpartition(sort_arr, 2))
```

```

# Import NumPy library
import numpy as np

# Create two 1D arrays for joining operations
j1 = np.array([1, 2])
j2 = np.array([3, 4])

# Create a 2x2 matrix for splitting operations
m_split = np.array([[1, 2], [3, 4]])

# Join two arrays into a single array
print('concatenate:', np.concatenate((j1, j2)))

# Stack arrays along a new axis (creates a 2D array)
print('\nstack:\n', np.stack((j1, j2)))

# Stack arrays vertically (row-wise)
print('\nvstack:\n', np.vstack((j1, j2)))

# Stack arrays horizontally (column-wise into 1D)
print('\nhstack:', np.hstack((j1, j2)))

# Stack arrays as columns into a 2D array
print('\ncolumn_stack:\n', np.column_stack((j1, j2)))

# Split array into equal parts
print('\nsplit:', np.split(np.array([1, 2, 3, 4]), 2))

# Split array into parts (unequal sizes allowed)
print('array_split:', np.array_split(np.array([1, 2, 3]), 2))

# Split matrix horizontally into two parts (column-wise)
print('\nhsplit:\n', np.hsplit(m_split, 2))

# Split matrix vertically into two parts (row-wise)
print('\nvsplit:\n', np.vsplit(m_split, 2))

# Repeat the entire array multiple times
print('\ntile:', np.tile(j1, 2))

# Repeat each element of the array multiple times
print('repeat:', np.repeat(j1, 2))

# Delete element at index 1
print('delete:', np.delete(j1, 1))

# Insert value 5 at index 1
print('insert:', np.insert(j1, 1, 5))

# Append one array to another
print('append:', np.append(j1, j2))

# Pad the array with zeros at the beginning and end
print('pad:', np.pad(j1, (1, 1), 'constant'))

```

## 9. Logical Ops

* `np.any()`: Any true?
* `np.logical_and()`: AND.
* `np.logical_or()`: OR.
* `np.logical_not()`: NOT.
* `np.greater()`: >
* `np.less()`: <
* `np.equal()`: ==
* `np.isnan()`: Is NaN?
* `np.isinf()`: Is Inf?
* `np.isclose()`: Close?

```
# Import NumPy library
import numpy as np

# Create Boolean arrays for logical operations
bool_a = np.array([True, False])
bool_b = np.array([True, True])

# Create an array containing normal value, NaN and Infinity
nan_inf = np.array([1, np.nan, np.inf])

# Check if all elements in bool_b are True
print('all:', np.all(bool_b))

# Check if any element in bool_a is True
print('any:', np.any(bool_a))

# Perform element-wise logical AND operation
print('logical_and:', np.logical_and(bool_a, bool_b))

# Perform element-wise logical OR operation
print('logical_or:', np.logical_or(bool_a, bool_b))

# Perform logical NOT operation
print('logical_not:', np.logical_not(bool_a))

# Check if first array elements are greater than second array
print('greater:', np.greater([2], [1]))

# Check if first array elements are less than second array
print('less:', np.less([1], [2]))

# Check if elements of two arrays are equal
print('equal:', np.equal([1], [1]))

# Check which elements are NaN (Not a Number)
print('isnan:', np.isnan(nan_inf))

# Check which elements are infinite
print('isinf:', np.isinf(nan_inf))

# Check if values are approximately equal within tolerance
print('isclose:', np.isclose([1e10], [1.0000000001e10]))
```

```
# Import NumPy library
import numpy as np

# Create a 2x2 matrix for linear algebra operations
m_linalg = np.array([[1, 2], [3, 4]])

# Create a vector for vector operations
v_linalg = np.array([1, 2])

# Perform matrix multiplication of m_linalg with itself
print('matmul:\n', np.matmul(m_linalg, m_linalg))

# Calculate the inverse of the matrix
print('\nlinalg.inv:\n', np.linalg.inv(m_linalg))

# Calculate the determinant of the matrix
print('\nlinalg.det:', np.linalg.det(m_linalg))

# Compute eigenvalues and eigenvectors of the matrix
vals, vecs = np.linalg.eig(m_linalg)

# Print eigenvalues
print('\nlinalg.eig vals:', vals)

# Calculate the vector norm (Euclidean length)
print('linalg.norm:', np.linalg.norm(v_linalg))

# Calculate the trace (sum of diagonal elements)
print('trace:', np.trace(m_linalg))

# Calculate inner product of vectors
print('inner:', np.inner(v_linalg, v_linalg))

# Calculate outer product of vectors
print('\nouter:\n', np.outer(v_linalg, v_linalg))

# Extract diagonal elements of the matrix
print('\ndiag:', np.diag(m_linalg))

# Calculate cross product of two 3D vectors
print('\ncross:', np.cross([1,0,0], [0,1,0]))
```

## 11. Random
* `np.random.rand()`: Uniform 0-1.
* `np.random.randn()`: Standard Normal.
* `np.random.randint()`: Random Int.
* `np.random.choice()`: Random Pick.
* `np.random.shuffle()`: Shuffle.
* `np.random.seed()`: Set seed.
* `np.random.normal()`: Normal dist.
* `np.random.uniform()`: Uniform dist.

```
# Import NumPy library
import numpy as np

# Set random seed so results are reproducible (same output every run)
np.random.seed(42)

# Generate 2 random numbers between 0 and 1 (uniform distribution)
print('rand:', np.random.rand(2))

# Generate 2 random numbers from standard normal distribution (mean=0, std=1)
print('randn:', np.random.randn(2))

# Generate 2 random integers between 0 and 9
print('randint:', np.random.randint(0, 10, 2))

# Randomly select 2 values from the given list
print('choice:', np.random.choice([1, 2, 3, 4], 2))

# Create an array to demonstrate shuffling
arr_shuff = np.array([1, 2, 3, 4, 5])

# Shuffle the array elements in-place (original array changes)
np.random.shuffle(arr_shuff)

# Print shuffled array
print('shuffle:', arr_shuff)

# Generate 2 random numbers from normal distribution (mean=0, std=1)
print('normal:', np.random.normal(0, 1, 2))

# Generate 2 random numbers from uniform distribution between 0 and 1
print('uniform:', np.random.uniform(0, 1, 2))
```
